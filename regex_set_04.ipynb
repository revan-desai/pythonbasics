```python
import re

s="united"

q=re.findall("[untd]",s)

print(q)

if q:
  print("match found")
else:
  print("not found")
```

```python
import re

s="python"

w=re.search("[a-d]",s)   # returns a match for any a lower case character between a and d  ie. a b c d

print(w)
```

```python
import re
s="portugal"
q=re.findall("[^qirk]",s)      # if above string contains atleast one character other than q,i,r,k then will return a match object

print(q)

if q:
  print("match found ")
else:
  print("not found")

```

```python
import re

s="54222"

q=re.findall("[a-zA-Z]",s)

if q:
  print("atleast one upper or lower case character found")
else:
  print("no upper or lower case character found")
```

```python
import re

s="i came here at 5:30PM and left at 9:45PM"

q=re.findall("[0-5][0-9]",s)   ##Check if the string has any two-digit numbers, from 00 to 59:

print(q)
```
