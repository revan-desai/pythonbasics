# metacharacter (^) --  Starts with     

```python
import re
s="Its raining today in pune"

p=re.search("^I",s)       # ^ starts with

if p:
  print("yes it starts with letter I ")
else:
  print("it does not starts with letter I ")

p1=re.search("^Z",s)

if p1:
  print("yes it starts with letter Z ")
else:
  print("it does not starts with letter Z ")
```

# metacharacter ($)-- ends with

```python
import re

s="Its raining today in pune"

q=re.search("e$",s)

if q:
  print("yes it ends with character e")
else:
  print("it does not end with character e")

q1=re.search("k$",s)

if q1:
  print("yes it ends with character k")
else:
  print("it does not end with character k")
```

# metacharacter (.)--any character-- if two dots then any two character

```python
s="computer"

re.search("c..p",s)

re.search("p...r",s)
```

# metacharacter (.*) -- zero or more characters

```python
import re

s="workplace"

k=re.search("w.*p",s)

print(k)

if k:
  print("yes match found")
else:
  print('match not found')

s="heo"                    #zero characters between he and o

k=re.search("he.*o",s)

print(k)
```

# metacharacter--(.+) --atleast one or more characters

```python
import re
s="notebook"

w=re.search("t.+b",s)

print(w)

s1="notbook"

w1=re.search("t.+b",s1)

print(w1)
```

# metacharacter (.?) --zero or one character only not more than one

```python
import re

s="flight"

q=re.search("g.?t",s)     # one character between g and t

print(q)

q1=re.search("f.?l",s)     # zero character between f and l

print(q1)

q2=re.search("f.?g",s)     # two characters between f and g     but we want zero or one hence result none

print(q2)

```
